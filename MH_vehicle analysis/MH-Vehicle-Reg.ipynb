Q2.  The file MH-Veh-Reg.csv contains data of vehicles registered in Maharashtra for the period 2000-2018.   
Write python code to do the following:

    i.Merge the data of each year (2000-2001, 2001-2002,….) to a single period (2000-2018) by the region and sub region.

    ii.Write a function that would take year as parameter and plot a graph. Classify the vehicles into 3 categories Private, Commercial and Others.  Commercial vehicles will have 2 subcategories Light Motor Vehicles (LMV) and Heavy Motor Vehicles (HMV).  Use different colors for each type of vehicle and also show the legend.  The graph should show both the values and percentages.

    Note:  
    I.Consider Motorcycles, Scooters, Moped, Cars, Jeeps, Stn. Wagons and Private Service Vehicles as Private and rest as Commercial.
    II.Bus, Truck, Trailer, Tractor, Tanker, Stage/Contract carriage are all HMV and rest LMV.

```python
def merge_data(df):

    df['Start_Year'] = df['Year'].str.split('-').str[0]
    df['Start_Year'] = pd.to_datetime(df['Start_Year'], format='%Y')
    end_year = pd.to_datetime('2018')

    def map_to_period(start_year):
        if start_year >= pd.to_datetime('2000') and start_year <= end_year:
            return "2000-2018"
        else:
            return "Other" 

    df['Period'] = df['Start_Year'].apply(map_to_period)

    selected_columns = ['Region', 'Sub Region', 'Period', 'Motor Cycles', 'Cars', 'Luxury /Turist Cabs/', 'Auto-rikshaws', 'Trucks & Lorries', 'Delivery Van (4 wheelers)', 'Delivery Van (3 wheelers)', 'Others']
    merged_df = df[selected_columns].groupby(['Region', 'Sub Region', 'Period']).sum().reset_index()

    return merged_df

merge_data(df).to_csv("merge.csv")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def classify_vehicles(data, year):
 
    private_vehicles = ['Motor Cycles', 'Scooters', 'Moped', 'Cars', 'Jeeps', 'Stn. Wagons', 'Private Service Vehicles']
    commercial_vehicles = ['Taxis meter fited', 'Luxury /Turist Cabs/', 'Auto-rikshaws', 'Stage carriages', 
                           'Contract carriages /Mini Bus', 'School Buses', 'Ambulances', 'Articulated/Multi.',
                           'Trucks & Lorries', 'Tanker', 'Delivery Van (4 wheelers)', 'Delivery Van (3 wheelers)',
                           'Tractors', 'Trailors']
    
    year_data = data[data['Start_Year'] == year]
    
    year_data[private_vehicles + commercial_vehicles + ['Others']] = year_data[private_vehicles + commercial_vehicles + ['Others']].astype(float)
    
    private_count = year_data[private_vehicles].sum().sum()
    commercial_count = year_data[commercial_vehicles].sum().sum()
    other_count = year_data['Others'].sum()
    
    lmv_count = year_data[commercial_vehicles[:-3]].sum().sum()
    hmv_count = commercial_count - lmv_count
    

    total_count = private_count + commercial_count + other_count
    
    if total_count == 0:
        print(f"No vehicle data available for the year {year}.")
        return
    
    # Calculate percentages
    private_percent = (private_count / total_count) * 100
    lmv_percent = (lmv_count / total_count) * 100
    hmv_percent = (hmv_count / total_count) * 100
    other_percent = (other_count / total_count) * 100
    
    # Plotting
    categories = ['Private', 'LMV', 'HMV', 'Others']
    counts = [private_count, lmv_count, hmv_count, other_count]
    percentages = [private_percent, lmv_percent, hmv_percent, other_percent]
    colors = ['blue', 'green', 'red', 'orange']
    
    plt.figure(figsize=(10, 6))
    plt.bar(categories, counts, color=colors)
    plt.xlabel('Vehicle Categories')
    plt.ylabel('Number of Vehicles')
    plt.title(f'Vehicle Classification for Year {year}')
    
    for i, (category, count, percent) in enumerate(zip(categories, counts, percentages)):
        plt.text(i, count + 100, f'{count} ({percent:.2f}%)', ha='center')
    
    plt.legend(['Private', 'LMV', 'HMV', 'Others'])
    plt.show()

# Example usage:
  # Replace 'your_file.csv' with the path to your CSV file containing the data
df['Start_Year'] = df['Year'].str.split('-').str[0]
df['Start_Year'] = pd.to_numeric(df['Start_Year'])

classify_vehicles(df, 2014)
```
